```python
import cv2 as cv
import matplotlib.pyplot as plt
import os
```

```python
# image path
image_path = '/content/img2.jpg'
```

```python
# loading image
img_original = cv.imread(image_path)
img_original = cv.cvtColor(img_original, cv.COLOR_BGR2RGB)

# showing the original image
plt.imshow(img_original),plt.title('Original')
```

```python
# convert image into image scale mode
img_gray_scale = cv.cvtColor(img_original, cv.COLOR_RGB2GRAY)

# showing the grayScale style the image
plt.imshow(img_gray_scale),plt.title('Gray Scale')
```

```python
# smoothing the colors of the image
img_blur = cv.medianBlur(img_gray_scale, 9)

# showing the grayScale style the image
plt.imshow(img_blur),plt.title('Blur')
```

```python
# detecting the edges
edges = cv.adaptiveThreshold(img_blur, 255, 
  cv.ADAPTIVE_THRESH_MEAN_C, 
  cv.THRESH_BINARY, 9, 2)

# showing the detected edges
plt.imshow(edges), plt.title('Edges')
```

```python
# creating a mask for the image
img_mask = cv.bilateralFilter(img_original, 9, 300, 300)

# showing mask of the image
plt.imshow(img_mask), plt.title('Mask')
```

```python
# convert a cartoonized image
img_cartoon = cv.bitwise_and(img_original, img_mask, mask=edges)

# showing the cartoonized image
plt.imshow(img_cartoon), plt.title('Cartoonized')
```
